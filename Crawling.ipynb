```python
for search_text in search_text_list[11:]:
    for length in search_period:
        begin_date = length[0]
        end_date = length[1]
        driver.get(f'https://patents.google.com/?q="{search_text}"&country=US&before=priority:{end_date}&after=priority:{begin_date}&language=ENGLISH')
        time.sleep(15)
        try:
            driver.find_element_by_xpath('//*[@id="count"]/div[1]/span[2]/a').click()
        except:
            continue
    download_folder = 'C:/Users/good5/Downloads/'
    file_list = os.listdir('C:/Users/good5/Downloads/')
    try:
        if not os.path.exists(f'./crawling_files/{search_text}/'):
            os.makedirs(f'./crawling_files/{search_text}/')
    except OSError:
        print(f"{search_text} Already Exist!")
    file_destination = f'./crawling_files/{search_text}/'
    for file in file_list:
        if file[:2]=='gp':
            shutil.move(download_folder + file, file_destination)
```

```python
code_dict = dict()
for search_text in search_text_list:
    path = f"./crawling_files/{search_text}/"
    file_list = os.listdir(path)
    code_list = []
    for file in file_list:
        df = pd.read_csv(f'./crawling_files/{search_text}/{file}', header=1)
        for i in range(len(df)):
            id_txt = df['id'][i]
            if 'B' in id_txt:
                code_list.append(id_txt.split('-')[1])
    code_dict[f'{search_text}'] = code_list
```

```python
import pickle

with open('code.pickle', 'wb') as fw:
    pickle.dump(code_dict, fw)
```

```python
with open('code.pickle', 'rb') as fr:
    code_loaded = pickle.load(fr)
```
